# Now we can create a Linear Regression model, using our description scores in place of the text!

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split 

train = pd.read_csv('clean_training_set.csv')
test = pd.read_csv('clean_test_set.csv')

target = train.Engagements
X = train.drop(columns=['Engagements','Description'])

for t_size in np.linspace(0.6,0.8,21):

    X_train, X_test, y_train, y_test = train_test_split(X, target, train_size = t_size, 
                                                        random_state=1) 
    # create linear regression object 
    reg = LinearRegression() 

    # train the model using the training sets 
    reg.fit(X_train, y_train);

    MAPE_Score = 1/len(y_test) * sum(abs(reg.predict(X_test) - y_test)/y_test)
    
    print(t_size, MAPE_Score)
```

Calculate the MAPE for the test data.

### Now use our model to fit the holdout_set.csv file.

```python
X_holdout = test.drop(columns=['Engagements','Description'])
engagement_prediction = reg.predict(X_holdout)
```

```python
## setting plot style 
plt.style.use('fivethirtyeight') 
  
## plotting residual errors in training data 
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, 
            color = "green", s = 1, label = 'Train data') 
  
## plotting residual errors in test data 
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, 
            color = "blue", s = 1, label = 'Test data') 
  
## plotting line for zero residual error 
plt.hlines(y = 0, xmin = 0, xmax = 50, linewidth = 2) 
  
## plotting legend 
plt.legend(loc = 'upper right') 
  
## plot title 
plt.title("Residual errors") 
  
## function to show plot 
plt.show() 
```
